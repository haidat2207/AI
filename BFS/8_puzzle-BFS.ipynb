##Bài code này sẽ tìm cách giải theo dạng BFS
##Bài này sẽ tìm cách nước đi có thể xảy ra và sau đó xét từng trường hợp nhỏ của các trường hợp đó cho đến khi tìm ra lời giải hoặc xét hết các trường hợp
##Người dùng chỉ cần nhập trạng thái của 8-puzzle thông qua ma trận 3x3

```python
from collections import deque


class Node:
    def __init__(self, state, node_parent=None, action=None, step=0):
        self.state = state
        self.parent = node_parent
        self.action = action
        self.step = step

def find_zero(node):

    for i in range(3):
        for j in range(3):

            if node[i][j] == 0:
                return (i, j)

    return None


def copy_node(node):

    return [row[:] for row in node]


def move(node):

    neighbors = []

    i, j = find_zero(node)

    directions = [
        (1, 0, "DOWN"),
        (-1, 0, "UP"),
        (0, 1, "RIGHT"),
        (0, -1, "LEFT")
    ]

    for di, dj, action in directions:

        new_i = i + di
        new_j = j + dj

        if 0 <= new_i < 3 and 0 <= new_j < 3:

            new_node = copy_node(node)

            new_node[i][j], new_node[new_i][new_j] = \
                new_node[new_i][new_j], new_node[i][j]

            neighbors.append((new_node, action))

    return neighbors


def BFS(initial_state, goal_state):

    root = Node(initial_state)

    frontier = deque([root])

    reached = []

    while frontier:

        current_node = frontier.popleft()

        if current_node.state == goal_state:
            return current_node

        reached.append(current_node.state)

        for neighbor, action in move(current_node.state):

            in_frontier = any(
                node.state == neighbor for node in frontier
            )

            if neighbor not in reached and not in_frontier:

                child = Node(
                    state=neighbor,
                    node_parent=current_node,
                    action=action,
                    step=current_node.step + 1
                )

                frontier.append(child)

    return None


def print_result(node):

    path = []

    while node:

        path.append((node.action, node.state, node.step))
        node = node.parent

    path.reverse()

    for action, state, step in path:
        print(f"Bước: {step}")
        print(f"Action: {action}")

        for row in state:
            print(row)

        print()


if __name__ == "__main__":

    goal_state = [
        [1, 2, 3],
        [4, 5, 6],
        [7, 8, 0]
    ]

    initial_state = []

    print("Nhập ma trận 3x3:")

    for i in range(3):
        row = list(map(int, input().split()))
        initial_state.append(row)

    result = BFS(initial_state, goal_state)

    if result:
        print_result(result)
        print("Đã tìm thấy giải pháp.")
    else:
        print("Không tìm thấy giải pháp.")
```
