## Bài giải random nước đi cho trò chơi 8-puzzle theo Model-based Agent
## Tập luật: Tìm ra những nước đi được sau đó kiểm tra xem trạng thái đó đã có trong quá khứ hay chưa để đưa vào những nước đi hợp lê.
## Bài sẽ random một trong những nước đi hợp lệ
## Bài sẽ kết thúc ghi đã giải xong hoặc không còn nước đi hợp lệ

```python
import random

goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

def print_puzzle(state):

    for row in state:
        print(row)

    print()

def find_zero(state):

    for i in range(3):
        for j in range(3):

            if state[i][j] == 0:
                return i, j


def rule(state, visited):

    i, j = find_zero(state)

    directions = [
        (1, 0, "DOWN"),
        (-1, 0, "UP"),
        (0, 1, "RIGHT"),
        (0, -1, "LEFT")
    ]

    valid_moves = []

    for di, dj, move_name in directions:

        new_i = i + di
        new_j = j + dj

        if 0 <= new_i < 3 and 0 <= new_j < 3:

            new_state = [row[:] for row in state]

            new_state[i][j], new_state[new_i][new_j] = \
                new_state[new_i][new_j], new_state[i][j]
            
            if new_state not in visited:
                valid_moves.append((new_state, move_name))

    if not valid_moves:
        return state, "NO VALID MOVES"
    
    return random.choice(valid_moves)

def action(initial_state):

    current_state = initial_state

    visited = []

    step = 0

    while True:

        # kết thúc nếu không còn nước đi hợp lệ
        if current_state in visited:

            print("KHÔNG CÒN NƯỚC ĐI HỢP LỆ!")
            return

        print("Bước", step)
        print_puzzle(current_state)

        # kiểm tra goal
        if current_state == goal_state:

            print("ĐÃ RA KẾT QUẢ!")
            return

        

        visited.append(current_state)

        # random move
        current_state, direction = rule(current_state, visited)

        print("Move:", direction)
        print("----------------")

        step += 1


if __name__ == "__main__":
    initial_state = []

    print("Nhập puzzle 3x3")

    for i in range(3):

        row = list(map(int, input().split()))

        initial_state.append(row)

    action(initial_state)
```
