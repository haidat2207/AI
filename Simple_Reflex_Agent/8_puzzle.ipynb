```python
import random

goal_state = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]

def print_puzzle(state):

    for row in state:
        print(row)

    print()

def find_zero(state):

    for i in range(3):
        for j in range(3):

            if state[i][j] == 0:
                return i, j

def move(state):

    i, j = find_zero(state)

    directions = [
        (1, 0, "DOWN"),
        (-1, 0, "UP"),
        (0, 1, "RIGHT"),
        (0, -1, "LEFT")
    ]

    valid_moves = []

    for di, dj, move_name in directions:

        new_i = i + di
        new_j = j + dj

        if 0 <= new_i < 3 and 0 <= new_j < 3:

            new_state = [row[:] for row in state]

            new_state[i][j], new_state[new_i][new_j] = \
                new_state[new_i][new_j], new_state[i][j]

            valid_moves.append((new_state, move_name))

    return random.choice(valid_moves)

def solve(initial_state):

    current_state = initial_state

    visited = []

    step = 0

    while True:

        print("Bước", step)
        print_puzzle(current_state)

        # kiểm tra goal
        if current_state == goal_state:

            print("ĐÃ RA KẾT QUẢ!")
            return

        # kiểm tra lặp
        if current_state in visited:

            print("ĐÃ LẶP LẠI TRẠNG THÁI CŨ!")
            return

        visited.append(current_state)

        # random move
        current_state, direction = move(current_state)

        print("Move:", direction)
        print("----------------")

        step += 1


initial_state = []

print("Nhập puzzle 3x3")

for i in range(3):

    row = list(map(int, input().split()))

    initial_state.append(row)

solve(initial_state)
```
