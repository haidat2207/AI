##Bài code này sẽ tìm cách giải theo dạng BFS,DFS theo 2 cách
##Bài này sẽ tìm cách nước đi có thể xảy ra và sau đó xét từng trường hợp nhỏ của các trường hợp đó cho đến khi tìm ra lời giải hoặc xét hết các trường hợp
##Người dùng chỉ cần nhập trạng thái của sàn nhà với bit 1 là sàn nhà dơ và bit 0 là sàn nhà sạch

```python
import tkinter as tk
from tkinter import simpledialog
import random

# =========================
# IMPORT THUẬT TOÁN
# =========================

from HutBui_BFS_1 import BFS, get_path
from HutBui_BFS_2 import BFS as BFS_2, get_path as get_path_2
from HutBui_DFS_1 import DFS, get_path as get_path_dfs
from HutBui_DFS_2 import DFS as DFS_2, get_path as get_path_dfs_2

# =========================
# CONSTANT
# =========================

CELL_SIZE = 60

# =========================
# WINDOW
# =========================

root = tk.Tk()

root.title("Máy hút bụi Visualizer")

# =========================
# NHẬP SIZE
# =========================

rows = simpledialog.askinteger(
    "Input",
    "Nhập số dòng:"
)

cols = simpledialog.askinteger(
    "Input",
    "Nhập số cột:"
)

# =========================
# NHẬP MA TRẬN
# =========================

matrix = []

for i in range(rows):

    row_data = simpledialog.askstring(
        "Input",
        f"Nhập dòng {i+1}:"
    )

    row = list(map(int, row_data.split()))

    matrix.append(row)

# =========================
# MAIN FRAME
# =========================

main_frame = tk.Frame(root)
main_frame.pack()

# =========================
# CANVAS
# =========================

canvas = tk.Canvas(
    main_frame,
    width=cols * CELL_SIZE,
    height=rows * CELL_SIZE,
    bg="white"
)

canvas.pack(side="left")

# =========================
# LOG BOX
# =========================

log_box = tk.Text(
    main_frame,
    width=35,
    height=25
)

log_box.pack(side="right")

# =========================
# DRAW MAP
# =========================

def draw_map(state):

    canvas.delete("all")

    for r in range(rows):
        for c in range(cols):

            x1 = c * CELL_SIZE
            y1 = r * CELL_SIZE
            x2 = x1 + CELL_SIZE
            y2 = y1 + CELL_SIZE

            # 1 = DIRTY
            if state[r][c] == 1:
                color = "red"

            # 0 = CLEAN
            else:
                color = "green"

            canvas.create_rectangle(
                x1,
                y1,
                x2,
                y2,
                fill=color,
                outline="black"
            )

# =========================
# ROBOT
# =========================

robot = None

# =========================
# PATH + STEP
# =========================

path = []

current_step = 0

# =========================
# START ALGORITHM
# =========================

def start_algorithm():

    global path
    global current_step

    current_step = 0

    selected_algorithm = algorithm_var.get()

    goal_state = [[0] * cols for _ in range(rows)]

    robot_x = random.randint(0, rows - 1)
    robot_y = random.randint(0, cols - 1)

    # BFS 1
    if selected_algorithm == "BFS 1":

        result = BFS(
            [row[:] for row in matrix],
            goal_state,
            robot_x,
            robot_y
        )

        path = get_path(result)

    # BFS 2
    elif selected_algorithm == "BFS 2":

        result = BFS_2(
            [row[:] for row in matrix],
            goal_state,
            robot_x,
            robot_y
        )

        path = get_path_2(result)

    # DFS 1
    elif selected_algorithm == "DFS 1":

        result = DFS(
            [row[:] for row in matrix],
            goal_state,
            robot_x,
            robot_y
        )

        path = get_path_dfs(result)

    # DFS 2
    elif selected_algorithm == "DFS 2":

        result = DFS_2(
            [row[:] for row in matrix],
            goal_state,
            robot_x,
            robot_y
        )

        path = get_path_dfs_2(result)

    log_box.insert(
        tk.END,
        f"Running {selected_algorithm}\n"
    )

    log_box.insert(
        tk.END,
        f"Robot start position: ({robot_x},{robot_y})\n\n"
    )

    log_box.see(tk.END)

# =========================
# NEXT ACTION
# =========================

def next_action():

    global current_step
    global robot

    if current_step >= len(path):
        return

    data = path[current_step]

    action = data["action"]
    state = data["state"]
    x = data["x"]
    y = data["y"]
    step = data["step"]

    # redraw map
    draw_map(state)

    # robot position
    x1 = y * CELL_SIZE + 10
    y1 = x * CELL_SIZE + 10

    x2 = x1 + 40
    y2 = y1 + 40

    robot = canvas.create_oval(
        x1,
        y1,
        x2,
        y2,
        fill="blue"
    )

    # update log
    log_box.insert(
        tk.END,
        f"Step {step}: {action} -> ({x},{y})\n"
    )

    log_box.see(tk.END)

    current_step += 1

# =========================
# DRAW INITIAL MAP
# =========================

draw_map(matrix)

# =========================
# BUTTON FRAME
# =========================

button_frame = tk.Frame(root)
button_frame.pack(side="bottom", pady=10)

# =========================
# ALGORITHM DROPDOWN
# =========================

algorithm_var = tk.StringVar(root)

algorithm_var.set("BFS 1")

algorithm_menu = tk.OptionMenu(
    button_frame,
    algorithm_var,
    "BFS 1",
    "BFS 2",
    "DFS 1",
    "DFS 2"
)

algorithm_menu.pack(side="left", padx=10)

# =========================
# START BUTTON
# =========================

start_btn = tk.Button(
    button_frame,
    text="Start",
    command=start_algorithm,
    width=12,
    bg="lightgreen"
)

start_btn.pack(side="left", padx=10)

# =========================
# NEXT BUTTON
# =========================

next_btn = tk.Button(
    button_frame,
    text="Next Action",
    command=next_action,
    width=12,
    bg="lightblue"
)

next_btn.pack(side="left", padx=10)

# =========================
# MAIN LOOP
# =========================

root.update()

root.mainloop()
```
